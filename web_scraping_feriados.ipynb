```python
import requests
from bs4 import BeautifulSoup
```

```python
URL_BASE = "https://feriados.cl/"

request_response = requests.get(URL_BASE)
```

```python
# a string
html_feriados = request_response.text
```

```python
# parseamos el html
soup = BeautifulSoup(html_feriados, "html.parser")
```

```python
# obtenemos la primera tabla
tabla = soup.find("table")
tabla
```

```python
# eliminamos todos los elementos con la clase "rl"
# estos elementos contienen las leyes, lo que no
# nos interesa
for element in tabla.find_all(class_="rl"):
    element.decompose()
```

```python
for element in tabla.find_all("span"):
    element.decompose()
```

```python
# obtenemos todos los tr
rows = tabla.find_all("tr")
print(rows)
```

```python
# no nos sirve la fila 0, 1, 2 y 3
feriados_rows = rows[4:]
# feriados_rows
```

```python
feriados_rows
```

```python
MONTHS = [
    "Enero", "Febrero", "Marzo", "Abril", "Mayo",
    "Junio", "Julio", "Agosto", "Septiembre",
    "Octubre", "Noviembre", "Diciembre"]

YEAR = "2024"

for feriado in feriados_rows:
    fecha = feriado.find("td")
    # print(fecha)
    # [nombre_dia, numero_dia, "de", nombre_mes]
    valores = fecha.get_text().split(" ")
    # print(valores)
    # buscamos que numero_mes es de acuerdo al nombre_mes
    numero_mes = MONTHS.index(valores[-1]) + 1
    # print(numero_mes)
    fecha.string.replace_with(f"{valores[1]}-{numero_mes}-{YEAR}")
    # print(fecha)
```

```python
feriados_rows
```

```python
lista_feriados = []

for f in feriados_rows:
    tds = f.find_all("td")

    datos = []
    for dato in tds:
        datos.append(dato.get_text(strip=True))

    lista_feriados.append(datos)

```

```python
lista_feriados
```

```python
# guardamos los datos en un csv
import csv

lista_feriados.insert(0, ["fecha","motivo","tipo"])
NOMBRE_ARCHIVO = "feriados.csv"

# escribimos los feriados en un archivo csv

with open(NOMBRE_ARCHIVO, "w") as archivo:
    escritor_csv = csv.writer(archivo)

    # escribimos cada feriado en una nueva línea
    for feriado in lista_feriados:
        print(feriado)
        escritor_csv.writerow(feriado)
```
